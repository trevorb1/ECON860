```python
fig, ax = plt.subplots(3)
ax[0].plot(df_population)
ax[1].plot(df_temp)
ax[2].plot(df_college)
ax[0].legend(df_population.columns.tolist(), loc='center left', bbox_to_anchor=(1,.5), prop={'size': 24})
ax[1].legend(df_temp.columns.tolist(), loc='center left', bbox_to_anchor=(1,.5), prop={'size': 24})
ax[2].legend(df_college.columns.tolist(), loc='center left', bbox_to_anchor=(1,.5), prop={'size': 24})
ax[0].set_title('City Population', fontsize = 32, pad=15)
ax[1].set_title('City Average High and Low Temperatures', fontsize = 32, pad=15)
ax[2].set_title('Proportion of College Aged Students', fontsize = 32, pad=15)
#ax[0].set_xlabel('Year', fontsize=24, labelpad=20)
ax[0].set_ylabel('Population',fontsize=24, labelpad=20)
ax[0].tick_params(axis = 'both', which = 'major', labelsize = 15)
ax[1].set_xlabel('Month', fontsize=20, labelpad=10)
ax[1].set_ylabel('Temperature (C)',fontsize=20, labelpad=10)
ax[1].tick_params(axis = 'both', which = 'major', labelsize = 15)
ax[2].set_xlabel('Year', fontsize=20, labelpad=10)
ax[2].set_ylabel('20-24 Year Olds (%)',fontsize=20, labelpad=10)
ax[2].tick_params(axis = 'both', which = 'major', labelsize = 15)
ax[2].set_ylim([0,20])

fig.set_size_inches(18, 28)
plt.show()
```
